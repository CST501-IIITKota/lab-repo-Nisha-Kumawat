# making most common type of plots using numpy 
## line , scatter, bar , hist , sublplot

# line

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(0,10,100)
x[:10]
fig,ax=plt.subplots(figsize=(2,2))
ax.plot(x,x**2)
plt.show()
```

# scatter

```python
fig,ax=plt.subplots(figsize=(3,3))
ax.scatter(x,np.exp(x));
#ax.set(title="scattered plot of exponential expression")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(3,3))
ax.scatter(x, np.sin(x));
plt.show()
```

# bar
verticle
horizontal

```python
nut={'amon':22,'peanut':57,'wallnut':11}
fig,ax= plt.subplots(figsize=(3,3))
ax.bar(nut.keys(),nut.values())
ax.set(title="Abhay's nut shop", ylabel="price(rupee)")
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(3,3))
ax.barh(list(nut.keys()),list(nut.values()))
plt.show()
```

# subplots : it uses line , bar , hist and scatter

```python
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                    ncols=2,
                                    figsize=(10,5))
#plot data on each axis:-
ax1.plot(x,x/2)
ax2.scatter(np.random.rand(10), np.random.rand(10))
ax3.bar(nut.keys(), nut.values())
ax4.hist(np.random.rand(1000))
plt.show()
```

```python
automobile=pd.read_csv("Automobile.csv")
automobile
```

```python
automobile['date']=pd.date_range('1/1/2025', periods=len(automobile))
automobile
```

```python
automobile['total_weight']=automobile['weight'].astype(int).cumsum()
automobile
```

# line

```python
automobile.plot(x='date', y='total_weight',)
plt.show()
```

# scatter

```python
automobile['acceleration']=automobile['acceleration'].astype(int)
automobile.plot(x='horsepower',y='acceleration',kind='scatter')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
heart_disease=pd.read_csv('heart_disease.csv')
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,5),subplots=True)
plt.show()
```

# ploting pandas using oo 
# ax.set_xlim()  works on a specific subplot (good for multiple subplots).
# plt.xlim()  works globally on whichever plot is currently active.

```python
over_70= heart_disease[heart_disease["Age"]>70]
over_70
```

```python
over_70.plot(kind='scatter' , x='Age',y='Cholesterol Level', c='Triglyceride Level', figsize=(5,6))
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,4))
over_70.plot(kind='scatter', x='Age',y='Cholesterol Level',c='Triglyceride Level', ax=ax)
ax.set_xlim([45,120])
plt.show()
```
