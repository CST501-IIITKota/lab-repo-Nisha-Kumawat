**Problem Statement 3: Sampling and the Central Limit Theorem (CLT**) **bold text**

Objective: To demonstrate the Central Limit Theorem by drawing samples from a non-normally distributed population.


1. Poopulation distribution:

```python
plt.figure(figsize=(10,5))
sns.histplot(df_cleaned['calculated_host_listings_count'], bins=50, kde=False)
plt.title("Distribution of Host Listings Count (Population)")
plt.xlabel("Number of Listings per Host")
plt.ylabel("Frequency")
plt.yscale("log")
plt.show()
```

```python
sample_means = []

for i in range(1000):
    sample = df_cleaned['calculated_host_listings_count'].sample(n=50, replace=True, random_state=None)
    sample_mean = sample.mean()
    sample_means.append(sample_mean)

sample_means = np.array(sample_means)

print("First 10 sample means:", sample_means[:10])
print("Average of sample means:", sample_means.mean())
```

```python
plt.figure(figsize=(10,5))
sns.histplot(sample_means, bins=30, kde=True)
plt.title("Sampling Distribution of Sample Means (n=50, 1000 samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean of Sample Means:", sample_means.mean())
```

**Explain how your findings demonstrate the Central Limit Theorem.**
**bold text**
If you repeatedly take random samples of size n from any population and compute their means, the distribution of those sample means will approach a normal distribution, regardless of the original populations shape.

**Problem Statement 4: Hypothesis Testing**

Objective: To use a t-test to determine if there is a statistically significant difference between two groups.


1. State the Hypotheses

Null Hypothesis (H):


Manhattan
	

=
Brooklyn
	


There is no difference in the average price of Airbnb listings between Manhattan and Brooklyn.

Alternative Hypothesis (H):

Manhattan

	

Brooklyn
	


There is a difference in the average price of Airbnb listings between Manhattan and Brooklyn.

This will be a two-tailed test.

```python
# Subset the data
manhattan_prices = df[df['neighbourhood_group'] == 'Manhattan']['price']
brooklyn_prices = df[df['neighbourhood_group'] == 'Brooklyn']['price']

# Quick check of sizes
print("Number of listings in Manhattan:", len(manhattan_prices))
print("Number of listings in Brooklyn:", len(brooklyn_prices))
```

```python
from scipy.stats import ttest_ind
# Set significance level
alpha = 0.05
ttest_result = ttest_ind(manhattan_prices, brooklyn_prices)

# Compare p-value to alpha
if ttest_result.pvalue < alpha:
    print(f"The p-value ({ttest_result.pvalue:.4f}) is less than the significance level ({alpha}).")
    print("We reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in the average price of Airbnb listings between Manhattan and Brooklyn.")
else:
    print(f"The p-value ({ttest_result.pvalue:.4f}) is greater than or equal to the significance level ({alpha}).")
    print("We fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in the average price of Airbnb listings between Manhattan and Brooklyn.")
```
