# Mean, Median, Mode

```python
data=[12, 15, 20, 20, 21, 22, 25, 25, 25, 30]

# mean 
mean=sum(data)/len(data)
# median
sorted_data= sorted(data)
n=len(data)
# ternary operator A if condition else B
median= (sorted_data[n//2] if n%2 != 0 else
        (sorted_data[n//2 - 1])+ sorted_data[n//2]/2)

#moode
mode=max(set(data), key=data.count)
mean, median , mode
```

Variance & Standard Deviation

```python
mean= sum(data)/len(data)
variance=sum((i-mean)**2 for i in data)/len(data)
sd=variance**0.2
variance, sd
```

5 number summary


```python
import numpy as np
import statistics as sts
data=[24,23,44,56,22,11,34,10,120]
q1,q2,q3= np.percentile(data,[25,50,75])
#print(q1,q2,q3)
iqr=q3-q1
lower_fence= q1-1.5*(iqr)
higher_fence=q3+1.5*(iqr)
print(q1,q2,q3,iqr,lower_fence,higher_fence)
```

measure of dispersion 

```python
sts.variance(data)
# it uses sample variance where we use n-1
sts.pvariance(data)
# it used to find population variance
```

```python
np.var(data,axis=0)
# it uses population variance where we uses n
```
